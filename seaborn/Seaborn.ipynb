##### FIRST LET US SEE ABOUT THE DIST PLOT


which deals with one variable

```python
sns.distplot(tips["total_bill"])
```

the wierd line is called KDE


and we can turn that off

```python
sns.distplot(tips["total_bill"],kde=False,bins=40)
```

##### distplots are used for analysing one variable


##### where as joint plots are used to compare two entities in a dataframe

```python
sns.jointplot(x='total_bill',y="tip",data=tips)
```

##### we can modify the graph in middle with te kind attribute

```python
sns.jointplot(x='total_bill',y="tip",data=tips,kind="hex")
```

### how the hex chart works is instead of plotting every point like in scatter graph the hex graph is made up of hexes if a point lies in hexagon then the color of that hexagon becomes darker

```python
sns.jointplot(x='total_bill',y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x='total_bill',y="tip",data=tips,kind="kde")
```

##### pairplot is like joint plot but joint plot to every set of columns in the dataframe

```python
sns.pairplot(tips)
```

#### the histograms are because x and y axis are same

```python
sns.pairplot(tips,hue="sex")
```

#### we can segreagate the columns in the chart with a attribute named hue
#### what is does is according to the values of the columns its gives u graphs

```python
sns.pairplot(tips,hue="size")
```

```python
sns.pairplot(tips,hue="size",palette="dark")
```

### rugplot is similar to distplot

 the difference between a rugplot and distplot is that the distplot involves the concept of bins and will add up all the data points in each bin, and plot this number where as the rugplot just plots a mark at each datapoint.





```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips["total_bill"],kde=False)
```

## grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris=sns.load_dataset("iris")
iris
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
```

```python
g=sns.PairGrid(iris)
g.map_upper(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_lower(sns.kdeplot)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
```
