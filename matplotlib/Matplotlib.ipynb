## legend

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
x=np.arange(1,11,2)
y=x*2
z=x*3
axes.plot(x,y,label="X Times Two")
axes.plot(y,z,label="X Times Three")
axes.legend()
```

#### for a legend to work we have to specify labels when we plot

Notice how are legend overlaps some of the actual plot!

The legend function takes an optional keyword argument loc that can be used to specify where in the figure the legend is to be drawn. The allowed values of loc are numerical codes for the various places the legend can be drawn. See the documentation page for details. Some of the most common loc values are:

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
x=np.arange(1,11,2)
y=x*2
z=x*3
axes.plot(x,y,label="X Times Two")
axes.plot(y,z,label="X Times Three")
axes.legend(loc=(0.6,0.1))
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple")
```

##### to change the color of the plot we add the attribute color

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=5)
```

##### we can also change the line width using the line width attribute which in default is 1

##### short hand notation is lw

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=5,alpha=0.2)
```

##### we can also change the line transparency using the alpha attribute which has values of 0-1

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=3,linestyle="--")
axes.plot(y,x,color="blue",linewidth=3,linestyle="-.")
axes.plot(x,x,color="orange",lw=3,ls=":")
```

##### using linestyle or ls we can change the plot style

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=2,marker="x")
```

##### markers are used to show where in the graph we have found the matches
##### markersize attribute is use d to increased marker size

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=2,marker="o",markersize=10
          ,markerfacecolor="yellow",markeredgewidth=3,markeredgecolor="green")
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple",linewidth=3,linestyle="--")
axes.plot(y,x,color="blue",linewidth=3,linestyle="-.")
axes.plot(x,x,color="orange",lw=3,ls=":")
axes.set_xlim([0,5])
axes.set_ylim([0,5])
```
